## Preparation of the data for visualization in Tableau purposes

```python
import pandas as pd
from collections import Counter
import seaborn as sns
import os
import numpy as np
from matplotlib import pyplot
import matplotlib.pyplot as plt
```

```python
# define the dataset location
filename = 'data/OnlineNewsPopularity.csv'

# Set Pandas to show all the columns
pd.set_option('display.max_columns', None)
# Read the data as a dataframe
data = pd.read_csv(filename)
```

```python
for column in data.columns:
    if column == ' self_reference_avg_sharess':
        data.rename(columns = {column:'self_reference_avg_shares'}, inplace = True)
    elif column == 'url':
        continue
    else:
        data.rename(columns = {column:column[1:]}, inplace = True)
```

```python
threshold = 1400 #higher than this is 1:popular
for i in range(data.shape[0]):
    val = data.loc[i,'shares']
    if val >= threshold:
        data.loc[i,'popular'] = 1
    else:
        data.loc[i,'popular'] = 0
```

```python
data.shape
```

```python
channels_cols = ['data_channel_is_entertainment', 'data_channel_is_bus', 'data_channel_is_socmed', 'data_channel_is_tech', 'data_channel_is_world']
days = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
channel = []
day = []
dropcols = []
for i in data.index:
    if any([data.loc[i][channels_cols[0]], 
            data.loc[i][channels_cols[1]], 
            data.loc[i][channels_cols[2]], 
            data.loc[i][channels_cols[3]], 
            data.loc[i][channels_cols[4]]])==False:
        channel.append(None) 
    else:
        for channel_ in channels_cols:
            if data.loc[i][channel_] == 1:
                l = channel_.split('_')
                channel.append(l[-1])
    for day_ in days:
        y = 'weekday_is_{}'.format(day_)
        if data.loc[i][y] == 1:
            day.append(day_)
        if y not in dropcols: dropcols.append(y)    
```

```python
dropcols += channels_cols
data = data.drop(dropcols, axis = 1)
```

```python
data['day'] = day
data['channel'] = channel
data
```

```python
data.to_csv('visualization_data.csv', index = False)
```

```python
data[data['popular']==1].to_csv('popular.csv', index = False)
data[data['popular']==0].to_csv('notpopular.csv', index = False)
```
